Basic NumPy Array Operations

```python
import numpy as np


arr = np.array([1, 2, 3, 4, 5])


print("Original array:", arr)
print("Mean:", np.mean(arr))
print("Standard Deviation:", np.std(arr))
print("Sum:", np.sum(arr))
print("Squared:", arr ** 2)
```

Plotting a Line Graph with Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='sin(x)', color='blue')
plt.title("Sine Wave")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

 Plotting a Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, color='green', alpha=0.4)
plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

Scatter Plot with Random Points

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, color='red', alpha=0.5)
plt.title("Random Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

Bar Chart Example

```python
import matplotlib.pyplot as plt
import numpy as np

languages = ['Python', 'Java', 'C++', 'JavaScript']
popularity = [85, 75, 65, 70]

x_pos = np.arange(len(languages))

plt.bar(x_pos, popularity, color='Blue')
plt.xticks(x_pos, languages)
plt.title("Programming Language Popularity")
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.show()
```

 Visualizing a 2D Sine Wave Heatmap

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of x and y values
x = np.linspace(-3*np.pi, 3*np.pi, 300)
y = np.linspace(-3*np.pi, 3*np.pi, 300)
X, Y = np.meshgrid(x, y)

# Compute Z as a function of X and Y
Z = np.sin(X) * np.cos(Y)

plt.imshow(Z, extent=[x.min(), x.max(), y.min(), y.max()], cmap='coolwarm', origin='lower')
plt.colorbar(label='Amplitude')
plt.title("2D Sine-Cosine Heatmap")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

Simple Line Plot of Squares

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 6)
y = x**2

plt.plot(x, y, marker='o')
plt.title("Square Numbers")
plt.xlabel("Number")
plt.ylabel("Square")
plt.show()
```

Pie Chart of Data Distribution

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [30, 25, 25, 20]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Category Distribution")
plt.axis('equal')
plt.show()
```
